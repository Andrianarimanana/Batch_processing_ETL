```python
def get_auxiliaire(text):
    if pd.isna(text):  # Si le texte dans fuil1 est NaN
        return 'NaN'
    for _, row in fuil2.iterrows():
        if pd.notna(row['Nom']) and row['Nom'] in text:  # Vérifie que 'Nom' n'est pas NaN
            return row['Auxiliaire']
    return 'NaN'

fuil1['Compte auxiliaire de contrepartie'] = fuil1['Nom_mapping'].apply(get_auxiliaire)
fuil1
```

```python
import pandas as pd

# DataFrames avec des valeurs possibles de type NaN
df1 = pd.DataFrame({
    'nom_mapping': ['apple pie', 'banana bread', 'cherry tart', 'blueberry muffin', None]
})

df2 = pd.DataFrame({
    'nom': ['apple', 'banana', 'cherry', 'grape'],
    'auxiliaire': ['apple1111', 'banana1323', 'cherry212', 'grape9999']
})

# Fonction qui gère les NaN
def get_auxiliaire(text):
    if pd.isna(text):  # Si la valeur est NaN, retourne KO
        return 'KO'
    for _, row in df2.iterrows():
        if row['nom'] in text:
            return row['auxiliaire']
    return 'KO'

# Appliquer la fonction pour créer la colonne auxiliaire
df1['auxiliaire'] = df1['nom_mapping'].apply(get_auxiliaire)

print(df1)
```

```python
import pandas as pd

# Exemple de données
fuil1 = pd.DataFrame({
    'Nom_mapping': ['apple pie', 'banana bread', 'cherry tart', 'blueberry muffin', None]
})

fuil2 = pd.DataFrame({
    'Nom': ['apple', 'banana', 'cherry', 'grape', None],
    'Auxiliaire': ['apple1111', 'banana1323', 'cherry212', 'grape9999', 'unknown']
})

# Fonction pour récupérer les valeurs correspondantes
def get_auxiliaire(text):
    if pd.isna(text):  # Si le texte dans fuil1 est NaN
        return 'KO'
    for _, row in fuil2.iterrows():
        if pd.notna(row['Nom']) and row['Nom'] in text:  # Vérifie que 'Nom' n'est pas NaN
            return row['Auxiliaire']
    return 'KO'

# Appliquer la fonction pour créer la colonne 'Auxiliaire'
fuil1['Auxiliaire'] = fuil1['Nom_mapping'].apply(get_auxiliaire)

print(fuil1)
```
